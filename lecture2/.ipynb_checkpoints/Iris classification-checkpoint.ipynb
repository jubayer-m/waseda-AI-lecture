```python
from sklearn.datasets import load_iris # declaration of the use of sciket-learn libray
iris_dataset = load_iris() # load iris dataset
print(iris_dataset.keys()) # show properties of the iris dataset
```

```python
print(iris_dataset.DESCR) # show the value of the property DESCR
```

```python
print(iris_dataset.target_names) # show the specy names
```

```python
print(iris_dataset.target) # show the classification outputs in the dataset
```

```python
print(iris_dataset.data) # show the inputs to classification
```

```python
print(iris_dataset.data[0])
```

```python
print(type(iris_dataset.data))  # data points are represented as `numpy.ndarray`
```

```python
# the `shape` property shows the dimensionality of the array and the number of the elements of each dimension
print(iris_dataset.data.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state=0)
```

```python
print(len(X_train), len(y_train), len(X_test), len(y_test)) # show the lengths of the split datasets
```

```python
print(X_train[0], y_train[0], X_test[0], y_test[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier  # load an implementation of K-nearest neighbors in scikit-learn (`sklearn`)
```

```python
K = 1
knn = KNeighborsClassifier(n_neighbors=K)
```

```python
knn.fit(X_train, y_train)  # train the K-nearest neighbors model
```

```python
# Build a data point classified by the `KNeighborsClassifier` model,
# which has to be 2-dimensional array
import numpy as np
x = [X_test[0]]
x = np.array(x)  # converts a list to a `nd.array`
print(x)
```

```python
knn.predict(x)
```

```python
y_test[0]
```

```python
X_test_predict = knn.predict(X_test)
```

```python
y_test == X_test_predict  # `nd.array` supports element-wise comparison
```

```python
# counts the number of the True elements
# True is converted to one, False is converted to zero, implicitly
np.sum(y_test == X_test_predict)
```

```python
accuracy = np.sum(y_test == X_test_predict) / y_test.shape[0]
print(accuracy)
```

```python
%matplotlib inline
import pandas as pd
df = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(df, c=iris_dataset.target, figsize=(15,15), marker='o', diagonal='')
```

```python
from mglearn.datasets import make_forge
X_forge, y_forge = make_forge()
```

```python
print(X_forge.shape, y_forge.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_forge, y_forge, random_state=7)
K = 1
knn = KNeighborsClassifier(n_neighbors=K)
knn.fit(X_train, y_train)
X_test_predict = knn.predict(X_test)
accuracy = np.sum(y_test == X_test_predict) / y_test.shape[0]
print(accuracy)
```

```python
df = pd.DataFrame(X_forge)
pd.plotting.scatter_matrix(df, c=y_forge, figsize=(5,5), marker='o', diagonal='')
```

```python
from sklearn.datasets import load_digits
digits_dataset = load_digits()
print(digits_dataset.data.shape, digits_dataset.target.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_dataset.data, digits_dataset.target, random_state=0)
K = 1
knn = KNeighborsClassifier(n_neighbors=K)
knn.fit(X_train, y_train)
X_test_predict = knn.predict(X_test)
accuracy = np.sum(y_test == X_test_predict) / y_test.shape[0]
print(accuracy)
```
